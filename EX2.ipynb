```python
df2=pd.read_csv(r'hcvdat0.csv')
```

```python
df2=df2.rename(columns={'Unnamed: 0':'Id'})
```

```python
df2.head()
```

```python
df2['Category'].unique()
```

```python
df2['Sex'].unique()
```

```python
df2['Category']=df2['Category'].map({'0=Blood Donor':'0','0s=suspect Blood Donor':'00','1=Hepatitis':'1','2=Fibrosis':'2','3=Cirrhosis':'3'})

df2['Sex']=df2['Sex'].map({'m':'0','f':'1'})
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.describe()
```

```python
df2['ALB']=df2['ALB'].fillna(df2['ALB'].mean())
```

```python
df2['ALP']=df2['ALP'].fillna(df2['ALP'].mean())
```

```python
df2['ALT']=df2['ALT'].fillna(df2['ALT'].mean())
```

```python
df2['CHOL']=df2['CHOL'].fillna(df2['CHOL'].median())
```

```python
df2['PROT']=df2['PROT'].fillna(df2['PROT'].median())
```

```python
df2.isnull().sum()
```

```python
df2.info()
```

```python
import seaborn as sns
plt.figure(figsize=(10,10))
c=df2.corr()
sns.heatmap(c,annot=True)
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x='CHOL',hue='Category',data=df2)
```

```python
sns.histplot(x='Category',hue='Category',data=df2)
```

```python
    df3=pd.read_csv(r'hepatitis_csv.xls')
    df3.head()
```

```python

df3.replace({
    'male': 0, 'female': 1,
    'False': 0, 'True': 1,
    False: 0, True: 1,
    'live': 1, 'die': 0
}, inplace=True)
```

```python
df3.head()
```

```python
df3.info()
```

```python
df3.isnull().sum()
```

```python
df3.describe()
```

```python
df3=df3.drop(columns='protime',axis=1)

```

```python
df3['liver_big']=df3['liver_big'].fillna(df3['liver_big'].mean())
df3['liver_firm']=df3['liver_firm'].fillna(df3['liver_firm'].mean())
df3['spleen_palpable']=df3['spleen_palpable'].fillna(0)
df3['spiders']=df3['spiders'].fillna(0)
df3['ascites']=df3['ascites'].fillna(0)
df3['varices']=df3['varices'].fillna(0)
df3['bilirubin']=df3['bilirubin'].fillna(df3['bilirubin'].mean())
df3['alk_phosphate']=df3['alk_phosphate'].fillna(df3['alk_phosphate'].mean())
df3['sgot']=df3['sgot'].fillna(df3['sgot'].mean())
df3['albumin']=df3['albumin'].fillna(df3['albumin'].mean())
```

```python
sns.boxplot(x='age',data=df3,hue='class')
```

```python
sns.scatterplot(x='alk_phosphate',y='sgot',data=df3,hue='class')
```
